# Add column with the text for each title

```python
def get_text(filename: str) -> str:
    with open(filename, 'r', encoding='utf-8') as f:
        content = f.read()
    return content
```

```python
df_full['text'] = df_full['local_filename'].map(get_text)
df_full.head()
```

# Remove metainformation

- At the beginning of each text there is name of the author and the name of the text. We already have this data in the dataframe, so we can ignore it now.
- At the end of each text there is metainformation about the text: who digitalized it, from where it was taken, etc. Remove it.

```python
def remove_suffix(content: str) -> str:
    idx_end = content.rfind('$id')
    return content[:idx_end]

def remove_prefix(content: str) -> str:
    idx_newline_fst = content.find("\n")
    idx_newline_snd = content.find("\n", idx_newline_fst + 1)
    return content[idx_newline_snd:]
```

## Suffix

```python
tmp = df_full['text'][0]
idx_end = len(tmp) - tmp.rfind('$id')
print(f'{idx_end=}')

print('################ Before ################')
print(df_full['text'][0][-idx_end-20:])
print('########################################')

df_full['text'] = df_full['text'].map(remove_suffix)
df_full.head()

print('################ After ################')
print(df_full['text'][0][-idx_end:])
print('########################################')
```

## Prefix

```python
print('################ Before ################')
print(df_full['text'][0][:100])
print('########################################')

df_full['text'] = df_full['text'].map(remove_prefix)
df_full.head()

print('################ After ################')
print(df_full['text'][0][:100])
print('########################################')
```
